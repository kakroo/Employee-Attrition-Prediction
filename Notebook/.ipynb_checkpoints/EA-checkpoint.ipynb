```python
#Plotting a bar chart

def bar_chart(feature):
    left= attritionDf[attritionDf['Attrition']=='Yes'][feature].value_counts()
    stayed= attritionDf[attritionDf['Attrition']=='No'][feature].value_counts()
    df = pd.DataFrame[left,stayed]
    #df.index=['Left Company','Stayed']
    df.plot(kind='bar')
```

```python
#Need to clean data

# Define a dictionary for the target mapping
target_map = {'Yes':1, 'No':0}
# Use the pandas apply method to numerically encode our attrition target variable
attritionDf["Attrition_numerical"] = attritionDf["Attrition"].apply(lambda x: target_map[x])
y=attritionDf["Attrition"].apply(lambda x: target_map[x])
target_map = {'Non-Travel':0, 'Travel_Rarely':1,'Travel_Frequently':2}
attritionDf["BusinessTravel"] = attritionDf["BusinessTravel"].apply(lambda x: target_map[x])
target_map = {'Sales':0, 'Research & Development':1,'Human Resources':2}
attritionDf["Department"] = attritionDf["Department"].apply(lambda x: target_map[x])
target_map = {'Other':0, 'Medical':1,'Marketing':2,'Life Sciences':3,'Technical Degree':4,'Human Resources':5}
attritionDf["EducationField"] = attritionDf["EducationField"].apply(lambda x: target_map[x])
target_map = {'Male':0, 'Female':1}
attritionDf["Gender"] = attritionDf["Gender"].apply(lambda x: target_map[x])
target_map = {'Single':0, 'Married':1,'Divorced':2}
attritionDf["MaritalStatus"] = attritionDf["MaritalStatus"].apply(lambda x: target_map[x])
target_map = {'No':0, 'Yes':1}
attritionDf["OverTime"] = attritionDf["OverTime"].apply(lambda x: target_map[x])



attritionDf.head()
```

```python
#Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```
